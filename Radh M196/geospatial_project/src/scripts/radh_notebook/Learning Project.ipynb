```python

class Person:
    def __init__(self, n, p, i):
        self.name = n
        self.personality = p
        self.is_sitting = i
        
    def sit_down(self):
        self.is_sitting = True
        
    def stand_up(self):
        self.is_sitting = False

# Objects

alice = Person("Alice", "aggressive", False)
backy = Person("Backy", "talkative", True)


class Robot:
# Constructer    
    def __init__(self, name, age, weight, master):
        self.master = master
        self.name = name
        self.age = age
        self.weight = weight
        
    def introduce_self(self):
        print("my name is " + self.name)
        
    def introduce_master(self):
        print("my master's name is " + self.master.name)
        
        

r1 = Robot("Radh", 20, 70, alice)
r2 = Robot("Rishov", 30, 75, backy)     

r1.introduce_self()
r1.introduce_master()

r2.introduce_self()
r2.introduce_master()
```
